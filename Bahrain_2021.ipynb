```python

import fastf1
from fastf1 import plotting
import matplotlib.pyplot as plt
import pandas as pd
import plotly.express as px


# Partindo dos seus DataFrames j√° carregados
laps_ham = race.laps.pick_driver('HAM')
laps_ver = race.laps.pick_driver('VER')

# Selecionando apenas as colunas de interesse
laps_ham_filtered = laps_ham[['LapNumber', 'Stint', 'LapTime', 'Compound','PitOutTime', 'PitInTime', 'TyreLife']]
laps_ver_filtered = laps_ver[['LapNumber', 'Stint', 'LapTime', 'Compound','PitOutTime', 'PitInTime', 'TyreLife']]

laps_ham_filtered['LapTimeSeconds'] = laps_ham_filtered['LapTime'].dt.total_seconds()
laps_ver_filtered['LapTimeSeconds'] = laps_ver_filtered['LapTime'].dt.total_seconds()

print(laps_ham_filtered)
print('--------------------')
print(laps_ver_filtered)
```

```python
laps_ham_filtered['LapTimeSeconds'] = laps_ham_filtered['LapTime'].dt.total_seconds()
laps_ver_filtered['LapTimeSeconds'] = laps_ver_filtered['LapTime'].dt.total_seconds()

media_stint1_ham = laps_ham_filtered[(laps_ham_filtered['LapNumber'] >= 1) & (laps_ham_filtered['LapNumber'] <= 13)]['LapTimeSeconds'].mean()
media_stint2_ham = laps_ham_filtered[(laps_ham_filtered['LapNumber'] >= 14) & (laps_ham_filtered['LapNumber'] <= 28)]['LapTimeSeconds'].mean()
media_stint3_ham = laps_ham_filtered[(laps_ham_filtered['LapNumber'] >= 29) & (laps_ham_filtered['LapNumber'] <= 56)]['LapTimeSeconds'].mean()

media_stint1_ver = laps_ver_filtered[(laps_ver_filtered['LapNumber'] >= 1) & (laps_ver_filtered['LapNumber'] <= 18)]['LapTimeSeconds'].mean()
media_stint2_ver = laps_ver_filtered[(laps_ver_filtered['LapNumber'] >= 19) & (laps_ver_filtered['LapNumber'] <= 40)]['LapTimeSeconds'].mean()
media_stint3_ver = laps_ver_filtered[(laps_ver_filtered['LapNumber'] >= 41) & (laps_ver_filtered['LapNumber'] <= 56)]['LapTimeSeconds'].mean()


print(f"Ritmo M√©dio do Hamilton (Voltas 1-13): {media_stint1_ham:.3f} segundos")
print(f"Ritmo M√©dio do Hamilton (Voltas 14-28): {media_stint2_ham:.3f} segundos")
print(f"Ritmo M√©dio do Hamilton (Voltas 29-56): {media_stint3_ham:.3f} segundos")

print(f"Ritmo M√©dio do Verstappen (Voltas 1-18): {media_stint1_ver:.3f} segundos")
print(f"Ritmo M√©dio do Verstappen (Voltas 19-40): {media_stint2_ver:.3f} segundos")
print(f"Ritmo M√©dio do Verstappen (Voltas 41-56): {media_stint3_ver:.3f} segundos")
```

```python
# Seus c√°lculos de m√©dia est√£o perfeitos. Vamos us√°-los.


diferenca_stint1 = abs(media_stint1_ham - media_stint1_ver)
vencedor_stint1 = "Hamilton" if media_stint1_ham < media_stint1_ver else "Verstappen"
print(f"No Stint 1, o piloto mais r√°pido foi {vencedor_stint1} por uma m√©dia de {diferenca_stint1:.3f} segundos.")

diferenca_stint2 = abs(media_stint2_ham - media_stint2_ver)
vencedor_stint2 = "Hamilton" if media_stint2_ham < media_stint2_ver else "Verstappen"
print(f"No Stint 2, o piloto mais r√°pido foi {vencedor_stint2} por uma m√©dia de {diferenca_stint2:.3f} segundos.")

diferenca_stint3 = abs(media_stint3_ham - media_stint3_ver)
vencedor_stint3 = "Hamilton" if media_stint3_ham < media_stint3_ver else "Verstappen"
print(f"No Stint 3, o piloto mais r√°pido foi {vencedor_stint3} por uma m√©dia de {diferenca_stint3:.3f} segundos.")
```

```python
import pandas as pd

# Encontre as voltas de pit stop para Hamilton e Verstappen (baseado em PitInTime)
pit_laps_ham = laps_ham[laps_ham['PitInTime'].notna()].reset_index(drop=True) # Reset index to easily access stops by number
pit_laps_ver = laps_ver[laps_ver['PitInTime'].notna()].reset_index(drop=True)


# --- AN√ÅLISE DA PRIMEIRA PARADA ---

print("-" * 20  )
print("An√°lise da Primeira Volta de Pit Stop:")

# Para Hamilton (primeira parada)
if len(pit_laps_ham) > 0:
    primeira_parada_ham_lap_data = pit_laps_ham.iloc[0]
    pit_lap_number_ham_1 = primeira_parada_ham_lap_data['LapNumber']
    pit_lap_time_ham_1 = primeira_parada_ham_lap_data['LapTime']
    print(f"Hamilton (Primeira Parada - Volta {pit_lap_number_ham_1}): {pit_lap_time_ham_1}" if pd.notna(pit_lap_time_ham_1) else f"Hamilton (Primeira Parada - Volta {pit_lap_number_ham_1}): Dados de tempo de volta indispon√≠veis")
else:
    pit_lap_number_ham_1 = 'N/A'
    pit_lap_time_ham_1 = pd.NA
    print("Hamilton (Primeira Parada): Sem dados de pit stop")

# Para Verstappen (primeira parada)
if len(pit_laps_ver) > 0:
    primeira_parada_ver_lap_data = pit_laps_ver.iloc[0]
    pit_lap_number_ver_1 = primeira_parada_ver_lap_data['LapNumber']
    pit_lap_time_ver_1 = primeira_parada_ver_lap_data['LapTime']
    print(f"Verstappen (Primeira Parada - Volta {pit_lap_number_ver_1}): {pit_lap_time_ver_1}" if pd.notna(pit_lap_time_ver_1) else f"Verstappen (Primeira Parada - Volta {pit_lap_number_ver_1}): Dados de tempo de volta indispon√≠veis")
else:
    pit_lap_number_ver_1 = 'N/A'
    pit_lap_time_ver_1 = pd.NA
    print("Verstappen (Primeira Parada): Sem dados de pit stop")


# Comparamos os tempos das primeiras voltas de pit stop (se dispon√≠veis)
if pd.notna(pit_lap_time_ham_1) and pd.notna(pit_lap_time_ver_1):
    diferenca_pit_lap_time_1 = abs(pit_lap_time_ham_1 - pit_lap_time_ver_1).total_seconds()
    vencedor_pit_lap_1 = "Hamilton" if pit_lap_time_ham_1 < pit_lap_time_ver_1 else "Verstappen"
    print(f"Conclus√£o da Primeira Parada: A volta de pit stop de {vencedor_pit_lap_1} foi {diferenca_pit_lap_time_1:.3f}s mais r√°pida.")
elif pd.notna(pit_lap_time_ham_1) or pd.notna(pit_lap_time_ver_1):
     print("N√£o foi poss√≠vel comparar as primeiras voltas de pit stop devido a dados ausentes para um dos pilotos.")
else:
    print("N√£o foi poss√≠vel comparar as primeiras voltas de pit stop devido a dados ausentes para ambos os pilotos.")


# --- AN√ÅLISE DA SEGUNDA PARADA ---

print("-" * 20  )
print("An√°lise da Segunda Volta de Pit Stop:")

# Para Hamilton (segunda parada)
if len(pit_laps_ham) > 1:
    segunda_parada_ham_lap_data = pit_laps_ham.iloc[1]
    pit_lap_number_ham_2 = segunda_parada_ham_lap_data['LapNumber']
    pit_lap_time_ham_2 = segunda_parada_ham_lap_data['LapTime']
    print(f"Hamilton (Segunda Parada - Volta {pit_lap_number_ham_2}): {pit_lap_time_ham_2}" if pd.notna(pit_lap_time_ham_2) else f"Hamilton (Segunda Parada - Volta {pit_lap_number_ham_2}): Dados de tempo de volta indispon√≠veis")
else:
    pit_lap_number_ham_2 = 'N/A'
    pit_lap_time_ham_2 = pd.NA
    print("Hamilton (Segunda Parada): Sem dados para a segunda parada")

# Para Verstappen (segunda parada)
if len(pit_laps_ver) > 1:
    segunda_parada_ver_lap_data = pit_laps_ver.iloc[1]
    pit_lap_number_ver_2 = segunda_parada_ver_lap_data['LapNumber']
    pit_lap_time_ver_2 = segunda_parada_ver_lap_data['LapTime']
    print(f"Verstappen (Segunda Parada - Volta {pit_lap_number_ver_2}): {pit_lap_time_ver_2}" if pd.notna(pit_lap_time_ver_2) else f"Verstappen (Segunda Parada - Volta {pit_lap_number_ver_2}): Dados de tempo de volta indispon√≠veis")
else:
    pit_lap_number_ver_2 = 'N/A'
    pit_lap_time_ver_2 = pd.NA
    print("Verstappen (Segunda Parada): Sem dados para a segunda parada")

# Comparamos os tempos das segundas voltas de pit stop (se dispon√≠veis)
if pd.notna(pit_lap_time_ham_2) and pd.notna(pit_lap_time_ver_2):
    diferenca_pit_lap_time_2 = abs(pit_lap_time_ham_2 - pit_lap_time_ver_2).total_seconds()
    vencedor_pit_lap_2 = "Hamilton" if pit_lap_time_ham_2 < pit_lap_time_ver_2 else "Verstappen"
    print(f"Conclus√£o da Segunda Parada: A volta de pit stop de {vencedor_pit_lap_2} foi {diferenca_pit_lap_time_2:.3f}s mais r√°pida.")
elif pd.notna(pit_lap_time_ham_2) or pd.notna(pit_lap_time_ver_2):
     print("N√£o foi poss√≠vel comparar as segundas voltas de pit stop devido a dados ausentes para um dos pilotos.")
else:
    print("N√£o foi poss√≠vel comparar as segundas voltas de pit stop devido a dados ausentes para ambos os pilotos.")
```

```python
all_laps = pd.concat([laps_ham, laps_ver])

all_laps_filtered = all_laps[['Driver', 'LapNumber', 'LapTime', 'Position']]

all_laps_filtered['LapTimeSeconds'] = all_laps_filtered['LapTime'].dt.total_seconds()
```

```python


fig_plotly = px.line(all_laps_filtered,
                     x='LapNumber',
                     y='LapTimeSeconds',
                     color='Driver',
                     title='An√°lise de Ritmo e Stints - Hamilton vs. Verstappen')

fig_plotly.show()

fig_plotly.write_html("bahrein_jp_plotly.html")

print("Gr√°fico interativo salvo como bahrein_jp_plotly.html")
```
